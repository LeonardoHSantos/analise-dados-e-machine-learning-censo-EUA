# Amostragem AleatÃ³ria Simples

### sample
A funÃ§Ã£o <b>"sample"</b> retorna valores aleatÃ³rios de um dataset, alguns parÃ¢metros podem ser passados para controlar a criaÃ§Ã£o de amostras de acordo com a necessidade, como exemplo:  <b>"n"</b>, <b>"replace"</b> e <b>"random_state"</b>.

### #n
NÃºmero de itens do eixo para retornar.

### #replace
Por padrÃ£o esse parÃ¢metro tem o valor <b>False</b>, se alterado para <b>True</b> retorna indices Ãºnicos na amostra aleatÃ³ria.
<br>replace = True --> nÃ£o repete os mesmos indices.

### #random_state
Caso este parametro receba um valor >= 1 serÃ¡ mantido os mesmos indices na amostra aleatÃ³ria.
<br>random_state = 1 --> mantem sempre os mesmo indices.


# Atente-se na diferenÃ§a!
Perceba a difereÃ§a entre repetiÃ§Ã£o e reuso. 
- <b>replace</b>: permite/nÃ£o permite os mesmos indices na amostra, por exemplo, se True nÃ£o haverÃ¡ repetiÃ§Ã£o do indice 1;
- <b>random_state</b>: permite/nÃ£o permite que sejam gerados novos indices, se >=1 serÃ¡ reutilizado os indices anteriores;



Esses sÃ£o apenas alguns parÃ¢mentros, para saber mais acesse a documentaÃ§Ã£o do Pandas:
https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.sample.html


# Evitando a repetiÃ§Ã£o de cÃ³digo

Seguindo as <b>boas prÃ¡ticas de codificaÃ§Ã£o</b>, irei implementar algumas funÃ§Ãµes para reutilizar durante a anÃ¡lise e deixar o cÃ³digo mais legÃ­vel.

Primeiramente irei criar uma funÃ§Ã£o para gerar uma amostra aleatÃ³ria simples usando os parÃ¢metros <b>n, replace e random_state</b>.

- dataset: a base com os dados;
- n (amostra): serÃ¡ a quantidade de amostra que desejamos;
- replace: definiÃ§Ã£o se haverÃ¡ indices exclusivos na amostra;
- random_state: define se queremos a repetiÃ§Ã£o dos indices todas as vezes que gerarmos amostras.





```python
def amostragem_aleatoria_simples(dataset, amostras, replace, random_state):
    return dataset.sample(n=amostras, replace=replace, random_state=random_state)
```

```python
df_amostra_aleatoria_simples = amostragem_aleatoria_simples(dataset, 100, True, 1)
df_amostra_aleatoria_simples
```

# Amostragem SistemÃ¡tica





```python
def amostragem_sistematica(dataset, n, r_seed=False):
    tam_dataset = len(dataset) // n
    
    if r_seed:
        random.seed(1)
    n_rand = random.randint(0, tam_dataset)
    
    ind_amostra_sist = np.arange(n_rand, len(dataset), step=tam_dataset)
    new_dataset = dataset.iloc[ind_amostra_sist]
    
#     print(tam_dataset, n_rand, amostra_sist)
    return new_dataset
```

```python
amostra_sist = amostragem_sistematica(dataset, 100, True)
amostra_sist
```
