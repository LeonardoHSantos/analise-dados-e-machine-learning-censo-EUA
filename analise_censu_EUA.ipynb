# Amostragem Aleatória Simples

### sample
A função <b>"sample"</b> retorna valores aleatórios de um dataset, alguns parâmetros podem ser passados para controlar a criação de amostras de acordo com a necessidade, como exemplo:  <b>"n"</b>, <b>"replace"</b> e <b>"random_state"</b>.

### #replace
Por padrão esse parâmetro tem o valor <b>False</b>, se alterado para <b>True</b> retorna indices únicos na amostra aleatória.
<br>replace = True --> não repete os mesmos indices.

### #random_state
Caso este parametro receba um valor >= 1 será mantido os mesmos indices na amostra aleatória.
<br>random_state = 1 --> mantem sempre os mesmo indices.

- <b>dataset</b>: a base com os dados;
- <b>n (amostra)</b>: será a quantidade de amostra que desejamos;
- <b>replace</b>: definição se haverá indices exclusivos na amostra;
- <b>random_state</b>: define se queremos a repetição dos indices todas as vezes que gerarmos amostras.




# Atente-se na diferença!
Perceba a difereça entre repetição e reutilização.
- <b>replace</b>: permite/não permite os mesmos indices na amostra, por exemplo, se True não haverá repetição do indice 1;
- <b>random_state</b>: permite/não permite que sejam gerados novos indices, se >=1 será reutilizado os indices anteriores;



Esses são apenas alguns parâmentros, para saber mais acesse a documentação do Pandas:
https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.sample.html


# Evitando a repetição de código

Seguindo as <b>boas práticas de codificação</b>, irei implementar algumas funções para reutilizar durante a análise e deixar o código mais legível.

Primeiramente irei criar uma função para gerar uma amostra aleatória simples usando os parâmetros <b>n, replace e random_state</b>.<br><br>


```python
def amostragem_aleatoria_simples(dataset, amostras, replace, random_state):
    return dataset.sample(n=amostras, replace=replace, random_state=random_state)
```

```python
df_amostra_aleatoria_simples = amostragem_aleatoria_simples(dataset, 100, True, 1)
df_amostra_aleatoria_simples
```

# Amostragem Sistemática

Quando precisamos gerar amostras aleatórias com maior nível de controle é comum o uso de funções auxiliares, para isso irei utilizar a biblioteca <b>numpy</b> do Python.

<hr>

### NUMPY - ARANGE

Lembrando que se houver dúvidas sobre a função <b>arange</b> segue link para a documentação:
<br>
https://numpy.org/doc/stable/reference/generated/numpy.arange.html

<br>

```python
def amostragem_sistematica(dataset, n, r_seed=False):
    # determinando a quantidade de itens para salto entre os indices do dataset
    tam_dataset = len(dataset) // n
    
    # se o valor de r_seed estiver como True a função "seed" manterá os mesmos indices a cada chamada.
    if r_seed:
        random.seed(1)
    
    # definição do inicio e fim do range de valores para a amostra.
    n_rand = random.randint(0, tam_dataset)
    
    indices_amostra_sist = np.arange(n_rand, len(dataset), step=tam_dataset)
    amostra_dataset = dataset.iloc[indices_amostra_sist]
    return amostra_dataset
```

```python
amostra_sist = amostragem_sistematica(dataset, 100, True)
amostra_sist
```

# Amostragem Por Grupos


O objetivo desse tipo de amostragem é adicionar caracteristicas para cada elemento, assim é possível obter maior nivel de profundidade na análise.

- Para o exemplo de amostras por agrupamento irei contruir a lógica separadamente para explicar alguns conceitos importantes sobre <b>checagem de valores</b>, após isso irei passar a lógica para uma função.

## tipo de uso para "grupos"

O uso mais comum desse tipo de amostragem (por grupos) é selecionar um grupo randomicamente e explorar os dados desse grupo de forma isolada.

### preparo da amostra
- 1) definir quais os tipos de grupos serão criados (quantidade de grupos);
- 2) selecionar randomicamente um grupo do dataset;

No o código a seguir, criarei uma nova função para gerar os grupos do dataset.

- <b>id_grupo</b>: armazena o do grupo;
- <b>cont_aux</b>: contagem de grupos;
- <b>grupos_dataset</b>: armazena os grupos do dataset;
- <b>quantidade_grupos</b>: quantidade de elementos no dataset dividido por 10.
<br><br>
- utilização do <b>loop for</b> possibilita atribuir valores dinamicamente as variavéis declaradas acima e controlar a quantidade de grupos.<br>
- a função do pandas "iterrows" auxilia o <b>loop for</b> a percorrer todas as linhas do dataset.

```python
id_grupo = 0
cont_aux = 0
grupos_dataset = list()

quantidade_grupos = len(dataset) // 10

for _ in dataset.iterrows():
    grupos_dataset.append(id_grupo)
    cont_aux += 1
    if cont_aux > quantidade_grupos:
        cont_aux = 0
        id_grupo += 1
```

# Convertendo os métodos de criação de grupos para função

Lembrando que uma das boas práticas de codificação é manter sempre seu código organizado, crie instruções dentro de funções para serem legiveis e reutilizaveis.

```python
def amostragem_por_grupos(dataset, n_grupos):
    id_grupo = 0
    cont_aux = 0
    grupos_dataset = list()

    quantidade_grupos = len(dataset) // n_grupos

    for _ in dataset.iterrows():
        grupos_dataset.append(id_grupo)
        cont_aux += 1
        if cont_aux > quantidade_grupos:
            cont_aux = 0
            id_grupo += 1
    
    dataset["grupo"] = grupos_dataset
    selecao_grupo_rand = random.randint(0, n_grupos)
    
    df_agrupamento = dataset[dataset["grupo"]==selecao_grupo_rand]
    return df_agrupamento
```

```python
df_agrupamento = amostragem_por_grupos(dataset, 22500)
df_agrupamento
```

# Amostra Estratificada

Este tipo de amostra é bastante utilizada quando precisamos analisar um percentual de uma população.

- exemplo: em uma população há 100 pessoas, precisamos analisar apenas os dados das mulheres (de forma simplificada temos que existam apenas duas opções: mulheres e homens).
- supondo que existam 60 mulheres e 40 homens na amostra.

Percebam que a quantidade de mulheres é maior do que a quantidade de homens, para amostras estratificadas devemos selecionar proporcionalmente a quantidade de elementos de forma que sejam equivalentes para ambos os conjuntos de dados, ou seja, é necessário obter uma amostra em percentual.

### Queremos analisar apenas 10% das mulheres e 10% dos homens.

#### mulheres
- 60 * 10 / 100 = 6,0

#### homens
- 40 * 10 / 100 = 4,0

Assim temos que deverá ser analisado <b>6 mulheres</b> e <b>4 homens</b>.

#### Em caso de valores quebrados, exemplo: 6,4.
Podemos aplicar a regra de arredondamento para menor ou maior ficando 6, caso seja >= 6,5 poderiamos arredondar para 7.

### Importante saber

Esse método serve para selecionar um subconjunto de dados, a análise deverá ser complementada por uma tecnica auxiliar como por exemplo, seleção randomica simples.

# amostra de treino e teste

Preparando amostras de forma inteligente...

Atualmente na área de dados é comum o uso de algoritimos de estatística e machine learning, já existem divesas bibliotecas desenvolvidas em Python com o objetivo de auxiliar na exploração de dados.

Um método bastante comum é a separação de amostras para treinar a máquina, após parametrizado é aplicado testes para validar os resultados.

### exemplos para dividir as amostras
- amostra para treino: 90%;
- amostra para teste: 10%.

A variavel <b>percentual_base</b> indicará o tamanho da amostra, no caso "0.1" representa 10% da base, será utilizado a função do sklearn <b>StratifiedShuffleSplit</b> para divisão da amostra treino e teste.

- percentual_base: tamanho da amostra;
- divisao: variavel que armazenará a função StratifiedShuffleSplit;
- df_x: deve representar 90% da base (amostra para treino);
- <b>df_y</b>: deve representar 10% da base <b>(amostra para teste)</b>;


```python
percentual_base = 0.1
divisao = StratifiedShuffleSplit(test_size=percentual_base)

for x, y in divisao.split(dataset, dataset["income"]):
    df_x = dataset.iloc[x]
    df_y = dataset.iloc[y]

df_x.shape, df_y.shape
```

Perceba a facilidade, com um loop for simples é possível definir os valores de teste e treino.

Mas podemos melhorar o código e para isso irei criar uma função chamada <b>amostra_estratificada</b> e deixar o código reutilizavél.

```python
def amostra_estratificada(base, test_size, coluna):
    percentual_base = test_size / len(dataset)
    divisao = StratifiedShuffleSplit(test_size=test_size)

    for x, y in divisao.split(dataset, dataset[coluna]):
        df_x = dataset.iloc[x]
        df_y = dataset.iloc[y]
    return df_x, df_y
```

# Uso da função de amostra estratificada

Pequenas mudanças foram realizada para deixar o código limpo e reutilizavel.

Basta informar os dados para função:
- base: nosso dataset;
- test_size: quantidade de registros para a amostra;
- coluna: nome da coluna que será a referência para função "split".

```python
amostra_estratif = amostra_estratificada(dataset, 100, "income")
```

```python
amostra_treino = amostra_estratif[0]
amostra_teste = amostra_estratif[1]

amostra_treino.shape, amostra_teste.shape
```
